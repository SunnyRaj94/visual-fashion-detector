In this notebook, we:
1. Load Grounding DINO.
2. Formulate query strings for fashion classes.
3. Run inference on street fashion.
4. Visualize using the interactive widget.

```python
import os
import sys
from typing import List, Dict
import numpy as np
import torch
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True

sys.path.append(os.path.abspath("."))

from fashion_detector.config import Config
from fashion_detector.models.grounding_dino import GroundingDinoDetector
from fashion_detector.models.fashion_clip import FashionClipDetector
from fashion_detector.utils import load_image
from fashion_detector.models.base import Detection

# Define parent category hierarchy mapping
CATEGORY_GROUPS = {
    "footwear": ["sneakers", "flats", "loafers", "mules slides", "heels", "sandals", "boots", "dress shoes"],
    "bag": ["tote bags", "shoulder bags", "crossbody bags", "handle bags", "backpacks", "belt bags", "clutches", "briefcases", "duffel bags", "messenger bags"],
    "upper body": ["tops", "sweaters", "shirts", "t shirts", "jackets blazers", "coats", "jackets", "blazers"],
    "lower body": ["pants", "jeans", "skirts", "shorts"],
    "one-piece": ["dresses", "suits sets", "jumpsuits", "suits"],
    "jewelry": ["jewelry", "earrings", "necklaces", "bracelets", "rings", "brooches"],
    "headwear": ["hats"],
    "eyewear": ["sunglasses"],
    "belt": ["belts"],
    "scarf": ["scarves shawls", "scarves"],
    "wallet": ["wallets"],
    "watch": ["watches"]
}

# Refined negative/neutral classes to filter out background, skin, face, hair, and nothing
NEGATIVE_CLASSES = ["human face", "hair", "skin", "background", "nothing"]

def compute_iou(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])
    inter = max(0, x2 - x1) * max(0, y2 - y1)
    area1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area2 = (box2[2] - box2[0]) * (box2[3] - box2[1])
    union = area1 + area2 - inter
    return inter / union if union else 0

def build_reverse_mapping() -> Dict[str, str]:
    mapping = {}
    for parent, children in CATEGORY_GROUPS.items():
        for child in children:
            mapping[child.lower().strip()] = parent
    return mapping

def run_hierarchical_dino_clip(image_path: str, user_cats: List[str], box_thresh: float = 0.25):
    config = Config("config/config.yaml")
    image = load_image(image_path)
    reverse_map = build_reverse_mapping()
    
    # 1. Map fine categories to parent queries
    parent_queries = set()
    for cat in user_cats:
        parent = reverse_map.get(cat.lower().strip())
        parent_queries.add(parent if parent else cat.lower().strip())
            
    parent_queries = list(parent_queries)
    print(f"Running Hierarchical Pipeline with box_threshold={box_thresh}")
    print(f"Parent queries: {parent_queries}\n")
    
    # 2. Stage 1: Grounding DINO proposals
    detector = GroundingDinoDetector(config)
    coarse_detections = detector.detect(image, queries=list(parent_queries), box_threshold=box_thresh)
    
    print(f"Grounding DINO returned {len(coarse_detections)} raw proposals.")
    
    # Sort detections by score descending before running NMS so we keep the most confident ones
    sorted_dets = sorted(coarse_detections, key=lambda d: d.score, reverse=True)
    
    # NMS
    filtered_coarse = []
    for det in sorted_dets:
        overlap = False
        for kept in filtered_coarse:
            if compute_iou(det.box, kept.box) > 0.45:
                overlap = True
                break
        if not overlap:
            filtered_coarse.append(det)
            
    print(f"Filtered down to {len(filtered_coarse)} unique boxes after NMS.")
    
    # 3. Stage 2: FashionCLIP classification (restricted to group + NEGATIVE_CLASSES)
    classifier = FashionClipDetector(config)
    classifier.load_model()
    
    final_detections = []
    w, h = image.size
    for det in filtered_coarse:
        parent = det.label.lower().strip()
        candidate_children = CATEGORY_GROUPS.get(parent, [])
        valid_candidates = [c for c in user_cats if c.lower().strip() in candidate_children]
        
        if not valid_candidates:
            valid_candidates = [det.label]
            
        classification_candidates = valid_candidates + NEGATIVE_CLASSES
        print(f"Crop '{parent}' (box: {list(map(int, det.box))}) will be classified against: {classification_candidates}")
        
        crop = image.crop((max(0, int(det.box[0])), max(0, int(det.box[1])), min(w, int(det.box[2])), min(h, int(det.box[3]))))
        
        # Get embeddings and classify
        text_features = classifier.get_text_features(classification_candidates)
        inputs = classifier.processor(images=crop, return_tensors="pt").to(classifier.device)
        with torch.no_grad():
            image_features = classifier.model.get_image_features(**inputs)
            image_features = image_features / image_features.norm(dim=-1, keepdim=True)
            similarity = (image_features @ text_features.T) * classifier.model.logit_scale.exp()
            probs = similarity.softmax(dim=-1).cpu().numpy()[0]
            
        best_idx = np.argmax(probs)
        best_label = classification_candidates[best_idx]
        best_score = float(probs[best_idx])
        
        # Filter out if matched to a negative/background class
        if best_label in NEGATIVE_CLASSES:
            print(f"  -> FILTERED OUT: Crop matched negative class '{best_label}' with confidence {best_score:.2f}")
            continue
            
        final_detections.append(Detection(
            box=det.box,
            label=best_label,
            score=best_score,
            metadata={"coarse_label": parent, "dino_score": det.score}
        ))
        
    print("\n--- Final Hierarchical Detections ---")
    for d in final_detections:
        print(f"- {d.label} (coarse: {d.metadata['coarse_label']}): dino_score={d.metadata['dino_score']:.2f}, clip_confidence={d.score:.2f}, box={list(map(int, d.box))}")
        
    return final_detections

if __name__ == "__main__":
    test_image = "/Users/sunnyraj/.gemini/antigravity-cli/brain/d4659731-b462-437a-91d5-3a41d61441ec/uploaded_image.jpg"
    user_categories = [
        'dresses', 'tops', 'sweaters', 'shirts', 'jackets blazers', 'coats', 'pants', 'jeans', 'skirts', 
        'suits sets', 'jumpsuits', 'sneakers', 'flats', 'loafers', 'mules slides', 'heels', 'sandals', 
        'boots', 'tote bags', 'shoulder bags', 'crossbody bags', 'handle bags', 'backpacks', 'belt bags', 
        'clutches', 'sunglasses', 'scarves shawls', 'belts', 'wallets', 'hats', 'watches', 'jewelry', 
        'earrings', 'necklaces', 'bracelets', 'rings', 'brooches', 't shirts', 'shirts', 'sweaters', 
        'jackets', 'blazers', 'coats', 'pants', 'jeans', 'suits', 'shorts', 'dress shoes', 'loafers', 
        'sneakers', 'boots', 'sandals', 'briefcases', 'backpacks', 'duffel bags', 'messenger bags', 
        'tote bags', 'sunglasses', 'belts', 'ties', 'scarves', 'wallets', 'hats', 'watches'
    ]
    user_categories = list(set(user_categories))
    
    if os.path.exists(test_image):
        final_detections = run_hierarchical_dino_clip(test_image, user_categories, box_thresh=0.25)
    else:
        print(f"Image not found at {test_image}")
```
